Build Neural Network

```python
# Import necessary packages
import skimage
import xml.etree.ElementTree as ET
import glob
import os
from pathlib import Path
import os
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import keras
from keras import layers
from sklearn.model_selection import train_test_split


# Define global variables
pap_images = glob.glob('/Users/ciarasikking/Data_Mining-Assignment3/Cropped/n02086910-papillon/*')
clum_images = glob.glob('/Users/ciarasikking/Data_Mining-Assignment3/Cropped/n02101556-clumber/*')
minp_images = glob.glob('/Users/ciarasikking/Data_Mining-Assignment3/Cropped/n02113712-miniature_poodle/*')
mexh_images = glob.glob('/Users/ciarasikking/Data_Mining-Assignment3/Cropped/n02113978-Mexican_hairless/*')

labels_pap = [0] * len(pap_images)
labels_clum = [1] * len(clum_images)
labels_minp = [2] * len(minp_images)
labels_mexh = [3] * len(mexh_images)

labels = labels_pap + labels_clum + labels_minp + labels_mexh
    
pap = []
for im in pap_images:
    read_image = plt.imread(im)
    pap.append(read_image)
    
clum = []
for im2 in clum_images:
    read_image2 = plt.imread(im2)
    clum.append(read_image2)

minp = []
for im3 in minp_images:
    read_image3 = plt.imread(im3)
    minp.append(read_image3)   

mexh = []
for im4 in mexh_images:
    read_image4 = plt.imread(im4)
    mexh.append(read_image4) 
    
data = pap + clum + minp + mexh

data_train, data_test, labels_train, labels_test = train_test_split(data, labels, random_state=104, test_size=0.20, shuffle=True)


#Build NN
num_classes = 4
input_shape = (128, 128, 3)
model = keras.Sequential(
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(8, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dense(16, activation="relu"),
        layers.Dense(num_classes, activation="softmax"),
    ]
)

model.summary()
```

```python
import tensorflow as tf
batch_size = 128
epochs = 15

labels_train_one_hot = tf.keras.utils.to_categorical(labels_train, num_classes=4)

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
history = model.fit(np.asarray(data_train), np.asarray(labels_train_one_hot), batch_size=batch_size, epochs=epochs, validation_split = 0.2)
```

```python
# Multiply Accuracies by 100
for key in history.history.keys():
    history.history[key] = [i * 100 for i in history.history[key]]
```

```python
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
